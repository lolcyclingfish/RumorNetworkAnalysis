```python
pol=pd.read_csv('data/politifact_clean.csv')
pol=pol[(~pd.isnull(pol['page_url'])) & (pol['page_url']!='about:blank')] #Get only online rows
```

```python
#f={'TRUE':'sum', 'page_url':'count'}
extra_features = dayOfWeek + tags
f = {i: 'sum' for i in extra_features}
f['TRUE']='sum'
f['page_url'] ='count'
pol_agg=pol.groupby(['Source','website'])['page_url','TRUE',
                                         'Monday', 'Tuesday', 'Wednesday', 
                                          'Thursday','Friday', 'Saturday', 'Sunday',
                                          'Elections', 'Health Care', 'Military', 'Religion', 'History'
                                         ].agg(f).reset_index().sort_values(by='page_url', ascending=False)
pol_agg['FALSE']=pol_agg['page_url']-pol_agg['TRUE']
pol=pol[pol['Source']!=pol['website']]
pol_agg.head()
```
